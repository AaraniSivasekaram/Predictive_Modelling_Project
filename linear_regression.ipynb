```python
# Import your dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Read in data
vaccine_df = pd.read_csv("Resources/vaccine_hesitancy_clean.csv")
vaccine_df.count()
```

```python
# Split dataset into X and y values
X = vaccine_df.drop(columns = ['estimated_hesitant'], axis = 1)
y = vaccine_df['estimated_hesitant']
```

```python
# Create an instance of the linear regression model 
model = LinearRegression()
```

```python
# Fit/training patterns in the model
model.fit(X,y)
```

```python
# Use predict() method to generate predictions
y_pred = model.predict(X)
print(y_pred)
```

```python
# Plot Residuals
plt.scatter(y_pred, y_pred - y, color="royalblue")
plt.hlines(y=0, xmin=y_pred.min(), xmax=y_pred.max(), color="red")
plt.title("Residual Plot")
plt.show()
```

```python
# Examine the slope (model.coef_) and y-intercept (model.intercept_)
print(model.coef_)
print(model.intercept_)
```

```python
# Plot predictions for each X column
import numpy as np
plt.scatter(X.iloc[:,0].values, y_pred)
x1 = vaccine_df.percent_hispanic
y1 = model.coef_[0]*x1+model.intercept_
plt.plot(x1, y1, color='red')
plt.xlabel('percent_hispanic')
plt.ylabel('Estimated Hesitancy')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
print('Mean Absolute Error: ', mean_absolute_error(y, y_pred))
print("Mean Squared Error: ", mean_squared_error(y, y_pred))
print("R Squared Score: ", r2_score(y, y_pred))
print("Explained Variance Score: ", explained_variance_score(y, y_pred))
```
